# 1) Let's first look at the Euclidean distance metric.









You are most likely to use Euclidean distance when calculating the distance between two rows of data that have numerical values, such a floating point or integer values.

> **Compute the Euclidean distance for one dimension.**
The distance between two points in one dimension is simply the absolute value of the difference between their coordinates.


```python
def EuclideanDistance1D(p1,p2):
  distance=abs(p1-p2)
  return distance
```

```python
EuclideanDistance1D(-3,5)
```


> **Compute the Euclidean distance for two and above dimensions.**

```python
import math 
def EuclideanDistance2D(p1,p2):
  distance=0
  for i,j in zip(p1,p2):
    distance += ((i-j)**2)
  return math.sqrt(distance)
```

# 2) Cosine Similarity





Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. **It is often used to measure document similarity in text analysis.**

```python
def dotProduct(a,b):
  dotProduct=0
  for (i, j) in zip(a, b):
    dotProduct += (i*j)
  return dotProduct
```

# 3) Correlation

Correlation analysis is used to quantify the degree to which two variables are related. Through the correlation analysis, you evaluate correlation coefficient that tells you how much one variable changes when the other one does. Correlation analysis provides you with a linear relationship between two variables.



> In statistics, correlation or dependence is any statistical relationship, whether causal or not, between two random variables or bivariate data.

Based on this formula, we need coveriance and standard deviation values.



> Standard Deviation:



In statistics, the **standard deviation** is a measure of the amount of variation or dispersion of a set of values.

StandartSapmaFormulu.png

```python
def standardDeviatoin (a):
  mean = sum(a) / len(a)
  total = 0
  for i in a:
    total += ((mean-i) ** 2)
    standardDev= math.sqrt(total/(len(a)-1))
  return standardDev  
```

```python
a=[1,2,3,4,5]
print("Standard Deviatoin:",standardDeviatoin(a))
```



> Coveriance:



In probability theory and statistics, **covariance** is a measure of the joint variability of two random variables.

covariance2.webp

```python
def coveriance(x,y):
  meanX = sum(x)/len(x)
  meanY = sum(y)/len(y)
  total=0
  for i,j in zip(x,y):
    total+=(meanX-i)*(meanY-j)
  return total/(len(x)-1)
```

```python
x=[5, 12, 18, 23, 45]
y=[2, 8, 18, 20, 28]
print("Coveriance:",coveriance(x,y))
```



>Let's calculate the **correlation**.



```python
def correlation(x,y):
  correlation = coveriance(x,y)/ (standardDeviatoin (x)*standardDeviatoin (y))
  return correlation
```

```python
x=[12,5,6,7,8]
y=[3,4,5,6,7]
print("Correlation:",correlation(x,y))
```

# 4) Manhattan Distance

We use Manhattan distance, also known as city block distance, or taxicab geometry if we need to calculate the distance between two data points in a grid-like path. 

For example 2 vectors like p and q

Taxicab geometry versus Euclidean distance:

In taxicab geometry, the red, yellow, and blue paths all have the same shortest path length of 12. In Euclidean geometry, the green line has length 8.49 and is the unique shortest path.


It is perhaps more useful to vectors that describe objects on a uniform grid, like a chessboard or city blocks. The taxicab name for the measure refers to the intuition for what the measure calculates: the shortest path that a taxicab would take between city blocks (coordinates on the grid).

```python
def manhattanDistance(a, b):
  distance = 0
  for i, j in zip(a,b):
    distance += abs(i-j) 
  return distance
```

```python
a = [10, 20,5]
b = [5, 5, 12]
manhattanDistance(a, b)
```

# 5) Minkowski Distance

Minkowski distance calculates the distance between two real-valued vectors.

It is a generalization of the Euclidean and Manhattan distance measures and adds a parameter, called the “order” or “p“, that allows different distance measures to be calculated

<img src = "https://wikimedia.org/api/rest_v1/media/math/render/svg/4ed8b780e0d3224880760b1745c444481590ee86" width=250 />

When p is set to 1, the calculation is the same as the Manhattan distance. When p is set to 2, it is the same as the Euclidean distance.

*   p=1: Manhattan distance.
*   p=2: Euclidean distance.





```python
def minkowskiDistance(a, b, p):
  total=0
  for e1, e2 in zip(a,b):
    total+=abs(e1-e2)**p
  total=total**(1/p)
  return  total   
```

```python

row1 = [5, 5, 5, 10, 5]
row2 = [0, 0, 5, 10, 0]
# calculate minkowski distance for (p=1)
minkowski = minkowskiDistance(row1, row2, 1)
print(minkowski)
# calculate minkowski distance for (p=2)
minkowski = minkowskiDistance(row1, row2, 2)
print(minkowski)
```

# 6) Hamming Distance



> **Hamming distance calculates the distance between two binary vectors, also referred to as binary strings or bitstrings for short.**





It is used for error detection or error correction when data is transmitted over computer networks. It is also using in coding theory for comparing equal length data words.

```python
def hammingDistance(a,b):
  total=0
  for e1, e2 in zip(a, b):
    total+= ((abs(e1-e2))/len(a))
  return total
```

```python
vector1=[0, 1, 1,  1]
vector2=[1, 1, 1,  1]
hammingDistance(vector1,vector2)
```

**Let's test our function using a library.**

```python
from scipy.spatial.distance import hamming
# define data
row1 = [0, 1, 1,  1]
row2 = [1, 1, 1,  1]
# print distance
print(hamming(row1, row2))
```
